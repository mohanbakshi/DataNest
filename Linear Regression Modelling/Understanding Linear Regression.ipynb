```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Preparing the data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) # Independent variable (hours studied)
y = np.array([2, 4, 5, 4, 5]) # Dependent variable (exam scores)
```

```python
# Create a linear regression model
model = LinearRegression()
# Train the model
model.fit(X, y)
```

```python
# Predict exam scores
y_pred = model.predict(X)
```

```python
# Plot the results
plt.figure(figsize=(8, 6))
# Plotting the actual data
plt.scatter(X, y, color='blue', label='Actual data')
# Plotting the best fit line
plt.plot(X, y_pred, color='red', label='Fitted line')
# Adding labels and title
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Linear Regression')
# Adding a legend
plt.legend()
# Show the plot
plt.show()
```

In the plot, the blue points represent the actual data, and the red line is the best fit line created by our linear regression model. This line helps us understand the relationship between hours studied and exam scores.
