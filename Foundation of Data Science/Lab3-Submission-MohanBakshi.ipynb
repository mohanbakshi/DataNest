<h3>Rearranging Data</h3>
<br>

```python
#Filtering data using boolean logic
filtered_data = edu[edu['TIME'] > 2005]
#pivot_table funtion is used to rearrange the data for better performance.
pivedu = pd.pivot_table(filtered_data, values='Value',
index=['GEO'], columns=['TIME'])
pivedu.head()
```

```python
#This operator allows to select specific rows.
pivedu.loc[['Spain', 'Portugal'], [2006, 2011]]
```

<br><br>
<h3>Ranking Data</h3>
<br>

```python
#Cleaning up the pivot table.
#Dropping Euro area enteries.
pivedu = pivedu.drop(['Euro area (13 countries)',
'Euro area (15 countries)',
'Euro area (17 countries)',
'Euro area (18 countries)',
'European Union (25 countries)',
'European Union (27 countries)',
'European Union (28 countries)'
], axis=0)

#Renaming the specific entry.
pivedu = pivedu.rename(
index={'Germany (until 1990 former territory of the FRG)': 'Germany'})

#Dropping the NaN values.
pivedu = pivedu.dropna()

#Ranking in the descending order.
pivedu.rank(ascending=False, method='first').head()
```

```python
#Making global ranking by summing up the columns using axis=1.
totalSum = pivedu.sum(axis=1)

#Ranking in the descending order.
totalSum.rank(ascending=False, method='dense').sort_values().head()
```

<h3>Plotting the Graphs based on Data</h3>

```python
from matplotlib import figure
from socket import socket

plt.rcParams.update(plt.rcParamsDefault)
# Creating a figure with a specific size
fig = plt.figure(figsize=(12, 5))

# Calculating and sorting the total values for each country
totalSum = pivedu.sum(axis=1).sort_values(ascending=False)

# Creating a bar chart with blue bars and 40% transparency
totalSum.plot(kind='bar', style='b', alpha=0.4,
title='Total Values for Country')

# Saving the plot as a PNG image with 300dpi resolution and tight borders
plt.savefig('Totalvalue_Country.png', dpi=300, bbox_inches='tight')
```

```python
#Defining colors for different years
my_colors = ['b', 'r', 'g', 'y', 'm', 'c']

#Plotting "Horizontal", "Stacked" bar graph using the defined colors
ax = pivedu.plot(kind='barh', stacked=True, color=my_colors, figsize=(12, 6))

#Setting the relative position of legend using "loc". "bbox_to_anchor" defines absolute position. 
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

#Saving figure.
plt.savefig('Value_Time_Country.png', dpi=300, bbox_inches='tight')


#Displaying figure.
plt.show()


```
