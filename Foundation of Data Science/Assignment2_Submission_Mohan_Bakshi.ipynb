# Plotting the Graphs

```python
# How many restaurants available per each city.
plt.figure(figsize=(8,6))
ax = sns.countplot(x='city',data=compelete_data,order = compelete_data['city'].value_counts().index)
plt.show()
```

#### We can observe Toronto has the maximum number of restaurants.

```python
# Reviews given user by city to restaurants.

compelete_data.groupby('city')['num_reviews'].sum().sort_values(ascending=False).plot(kind='pie')
plt.title("Number of reviews per city", fontsize=20)
plt.show()
```

#### We can observe Vancouver & Toronto city users gives more reviews to restaurant compare to other cities.

```python
# Generate skewed data
skewed_data = compelete_data['star']
# Create the density plot with labels
plt.figure(figsize=(5, 5)) # Set figure size
density_plot = skewed_data.plot(kind="density", label="Skewed Distribution")

# Plot mean and median lines with labels
mean_line = plt.vlines(skewed_data.mean(), ymin=0, ymax=0.8, linewidth=5.0, color="black", label="Mean")
median_line = plt.vlines(skewed_data.median(), ymin=0, ymax=0.8, linewidth=5.0, color="red", label="Median")

# Create proxy artist
proxy = plt.plot([], [], label="Skewed Distribution", linestyle="-"
, color="blue")[0]

# Add legends
plt.legend(handles=[proxy, mean_line, median_line])

# Set axis limits and labels
plt.xlim(-1, 5)
plt.xlabel("Value")
plt.ylabel("Density")

# Display the plot
plt.title("Density Plot of Skewed Stars given by user with Mean and Median Lines")
plt.show()
```
