# Constructing Data

`The requirement for this assignment is to derice atleast 5 attributes.`

In the section, we will be transforming our datasets and deriving new attributes.

The attributes we are going to derive are:

`1. Average Rating per Review (star / num_reviews):`
This attribute calculates the average star rating per review for each restaurant. It provides insights into the consistency of a restaurant's ratings. Higher values might indicate a more consistent experience, while lower values could suggest a wider range of customer experiences.

`2. Category Combination (category_1 + ' - ' + category_2):`
Combine the category_1 and category_2 columns into a single string with a delimiter (e.g., hyphen). This simplifies analysis by grouping restaurants with similar cuisine types (e.g., Italian - Pizza).

`3. City Popularity (num_reviews per City):`
Calculate the average number of reviews per city (total reviews for the city divided by the number of restaurants in that city). This attribute highlights relatively popular cities based on review volume.

`4. Distance Category (e.g., Close, Medium, Far):`
This categorizes restaurants based on their distance from a reference point (e.g., user location, city center). We can define thresholds for "close," "medium," and "far" based on the data distribution or domain knowledge (average travel times in the city).

`5. High Review Volume Indicator:`
This creates a binary feature to identify potentially well-established or popular restaurants based on a defined threshold for the number of reviews.

### 1. Deriving Average Rating per Review (star / num_reviews) column


```python
# Handle potential null values in num_reviews
doordash_cities_df['avg_rating_per_review'] = doordash_cities_df['star'] / doordash_cities_df.get('num_reviews', 0)  

doordash_cities_df.head(2)
```

### 2. Deriving Category Combination (category_1 + ' - ' + category_2) column

```python
doordash_cities_df['category_1'] = doordash_cities_df['category_1'].astype(str)
doordash_cities_df['category_2'] = doordash_cities_df['category_2'].astype(str)
doordash_cities_df['category_combo'] = doordash_cities_df['category_1'] + ' - ' + doordash_cities_df['category_2']

doordash_cities_df.head()
```

### 3. Deriving City Popularity (num_reviews per City) column

```python
# Grouping doordash_cities_df by 'city' and calculating the sum of 'num_reviews' for each city
city_reviews = doordash_cities_df.groupby('city')['num_reviews'].sum()

# Grouping doordash_cities_df by 'city' and counting the number of occurrences of each city 
city_counts = doordash_cities_df.groupby('city')['city'].count()

# Aggregating data using the agg() function:
#   - 'total_reviews_per_city': Calculating the sum of 'num_reviews' for each city using 'sum' aggregation
#   - 'city_count': Counting the number of occurrences of each city using 'count' aggregation
# Reseting the index to obtain a regular DataFrame after grouping

city_data = doordash_cities_df.groupby('city').agg(total_reviews_per_city=('num_reviews', 'sum'), city_count=('city', 'count')).reset_index()

# Merging the original doordash_cities_df with the calculated city data (city_data)
#   - on='city': Merge based on the 'city' column in both DataFrames
#   - how='left': Keep all rows from doordash_cities_df and matching rows from city_data

doordash_cities_df = doordash_cities_df.merge(city_data, on='city', how='left')

# Calculating the 'city_popularity' as the ratio of total reviews per city to the city count
doordash_cities_df['city_popularity'] = doordash_cities_df['total_reviews_per_city'] / doordash_cities_df['city_count']

doordash_cities_df
```

### 4. Deriving Distance Category (e.g., Close, Medium, Far) column

```python
distance_categories = {
  'Close': (0, 2),  # Distance in relevant unit km/miles (or relevant unit)
  'Medium': (2, 5),
  'Far': (5, np.inf)  # Infinity for upper bound
}

# Extracting lower and upper bounds from the dictionary values
bins = [val[0] for val in distance_categories.values()] + [np.inf]  # Adding infinity for the upper bound of the last bin

doordash_cities_df['distance_category'] = pd.cut(doordash_cities_df['distance'], bins=bins, labels=list(distance_categories.keys()))
doordash_cities_df.head(2)
```

### 5. Deriving High Review Volume Indicator column

```python
# Setting a threshold for high review volume
high_review_threshold = 50

doordash_cities_df['high_review_volume'] = np.where(doordash_cities_df['num_reviews'] >= high_review_threshold, 1, 0)

doordash_cities_df.head(2)
```

# Plotting

```python
# Calculating average rating per city
avg_rating_by_city = doordash_cities_df.groupby('city')['avg_rating_per_review'].mean()

# Creating the bar chart
plt.figure(figsize=(6, 6))
plt.bar(avg_rating_by_city.index, avg_rating_by_city.values, label='Distribution')
plt.xlabel('City')
plt.ylabel('Average Rating per Review')
plt.title('Average Rating Distribution by City')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```

This bar chart, titled <b>Average Rating Distribution by City</b>, depicts the average rating per review for various cities in your Doordash data. Here is what we can observations:

The bar heights represent the average rating each city receives. Cities like `Vancouver` has the highest average rating per review aprox. ~.200 which means it has the highest average customer satisfaction with restaurants, while cities like `Winnipeg` and `Calgary` have comparatively a lower average rating.

```python
# Calculating city popularity (total reviews divided by number of restaurants per city)
doordash_cities_df['city_popularity'] = doordash_cities_df['num_reviews'] / doordash_cities_df.groupby('city')['city'].transform('count')

# Grouping by city and sum city popularity for pie chart (to avoid duplicate city entries)
city_popularity_sum = doordash_cities_df.groupby('city')['city_popularity'].sum()

# Creating the pie chart
plt.figure(figsize=(4, 4))
plt.pie(city_popularity_sum, labels=city_popularity_sum.index, autopct='%1.1f%%', startangle=140)  # Adjust startangle for better label placement

# Adding title
plt.title('City Popularity Distribution (Avg. Reviews per City)\n')

# Equal aspect ratio ensures a circular pie chart
plt.axis('equal')

plt.show()
```

Based on the pie chart titled <b>City Popularity Distribution (Avg. Reviews per City)</b> we can observer that:

<b>Dominant City:</b> Vancouver has the highest average review volume (36.0%) among the three cities, indicating it's the most popular based on the number of reviews per restaurant.

<b>Close Competition:</b> Winnipeg (32.3%) and Calgary (31.8%) have relatively close average review volumes, suggesting they might be competing for second place in terms of city popularity.

`Overall, the pie chart highlights Vancouver as the leader in average review volume, followed by a close competition between Winnipeg and Calgary.`
