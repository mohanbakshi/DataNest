```python
# classify columns into categorical and quantitative variables
def check_variables(df: pd.DataFrame):
    cols = df.columns
    quantitative_cols = df._get_numeric_data().columns
    categorical_cols = list(set(cols) - set(quantitative_cols))
    return categorical_cols, list(set(quantitative_cols))

cat_cols, quan_cols = check_variables(df_insurance)
print(f'categorical columns: {cat_cols}')
print(f'quantitative columns: {quan_cols}')
```

# Examine the relationships between categorical explanatory variables and quantitative response variables from the dataset, a pair (C->Q) at a time using
- Descriptive statistics
- Side-by-side box plots
- Interpret your plots and calculated statistics

```python
# Function that will examine the relationshionship between categorical explanatory variables and quantitative response variables 
# through the Descritive statistics and sisde-by-side box plots.

def categorical_to_quantitative(df, categorical_vars, quantitative_var):
        print(f"\nExamination of relationship between{categorical_vars} and {quantitative_var}:")
        
        # Descriptive statistics
        grouped_data = df.groupby(categorical_vars)[quantitative_var].describe()
        print(grouped_data)
        
        # Side-by-side box plots
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=categorical_vars, y=quantitative_var, data=df)
        plt.title(f"{categorical_vars} vs {quantitative_var}")
        plt.xlabel(categorical_vars)
        plt.ylabel(quantitative_var)
        plt.show()
```

# Examine the relationship between categorical variables from the dataset, a pair (C->C) at a time, using:
- two-way tables as from lecture
- conditional percentages and double bar charts
- Interpret your plots and calculation

```python
def categorical_to_categorical(df, cat_var1, cat_var2):
    # Create a two-way table
    two_way_table = pd.crosstab(df[cat_var1], df[cat_var2])
    print("\nTwo-Way Table:")
    print(two_way_table)
    
    # Calculate conditional percentages
    conditional_percentages = two_way_table.div(two_way_table.sum(axis=1), axis=0) * 100
    
    print("\nConditional Percentages:")
    print(conditional_percentages)
    
    # Plot double bar charts
    labels = list(two_way_table.index)
    width = 0.4
    x = np.arange(len(labels))
    
    bar_labels = list(df[cat_var2].unique())  # Get unique elements of cat_var2
    
    fig, ax = plt.subplots(figsize=(10, 6))
    rects = []
    for i, label in enumerate(bar_labels):
        bar_percentage = conditional_percentages[label]
        rects.append(ax.bar(x + (-0.5 + i) * width / len(bar_labels), bar_percentage, width / len(bar_labels), label=label))
    
    ax.set_ylabel('Percent')
    ax.set_xlabel(cat_var1)
    ax.set_title(f'Double Bar Chart: {cat_var1} vs {cat_var2}')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    for rect in rects:
        ax.bar_label(rect, padding=3)
    plt.show()
```

```python
categorical_to_categorical(df_insurance,'sex','smoker')
```

# Examine the relationship between quantitative variables, a pair (Q->Q) at a time, using
- scatter plots for reporting direction, form, strength, and outliers.
- calculate linear correlation
- calculate least square regression line. Plot the regression line within thescatter plots.
- Interpret your plots and calculation

```python
def quantitative_to_quantitative(df, var1, var2):
    # Calculate linear correlation
    corr_mat = np.corrcoef(df[var1], df[var2])
    r = corr_mat[0][1]
    print("\nLinear Correlation:")
    print(corr_mat)
    
    # Scatter plot for two quantitative variables
    plt.figure(figsize=(10, 6))
    plt.scatter(df[var1], df[var2])
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.title(f'Scatter Plot: {var1} vs {var2}, r = {round(r, 4)}')
    plt.show()
    
    # Scatter plot with least square regression line
    # Calculate mid value of var2
    mid_value = np.median(df[var2])
    # Filter data for charges less than mid value of var2
    df_small = df[df[var2] <= mid_value]
    # Calculate linear correlation
    r_small = np.corrcoef(df_small[var1], df_small[var2])[0][1]
    # Calculate slope
    b = r_small * (np.std(df_small[var2]) / np.std(df_small[var1]))
    # Calculate intercept
    a = np.mean(df_small[var2]) - b * np.mean(df_small[var1])
    # Regression line
    x = np.linspace(min(df_small[var1]), max(df_small[var1]), 100)
    y = a + b * x
    # Plot scatter
    plt.figure(figsize=(10, 6))
    plt.scatter(df_small[var1], df_small[var2], label=f'{var2} <= {mid_value}')
    # Plot regression line
    plt.plot(x, y, '-r', label=f'{var2} = {round(a, 3)} + {round(b, 3)} * {var1}')
    plt.xlabel(var1, color='#1C2833')
    plt.ylabel(var2, color='#1C2833')
    plt.legend(loc='upper left')
    plt.title('Least Square Regression')
    plt.show()
```

```python
quantitative_to_quantitative(df_insurance,'age','charges')
```
